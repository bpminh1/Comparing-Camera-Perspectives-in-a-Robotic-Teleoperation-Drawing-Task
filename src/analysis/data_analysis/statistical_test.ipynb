```python
from load_data import DataLoader
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
from scipy.stats import wilcoxon, binomtest, spearmanr, gaussian_kde, kurtosis, skew, chi2_contingency
```

```python
loader = DataLoader()
loader.load_data()
df = loader.data
df_questionnaire = loader.questionnaire
df_questionnaire_preferences = loader.questionnaire_preferences
```

```python
def perform_t_test(result, matrix_len, value_map):
    R = np.zeros((1, matrix_len))
    for key, value in value_map.items():
        R[0, result.model.exog_names.index(key)] = value
    display(result.t_test(R))
```

**MODELS**

```python
model_trade_off = smf.mixedlm(
    "log_c ~ log_d * camera", 
    data=df, 
    groups=df["participant"], 
    re_formula="~camera"
)

result_trade_off = model_trade_off.fit() 
print(result_trade_off.summary())
```

```python
model_accuracy = smf.mixedlm(
    "accuracy ~ camera",
    data=df[df['block'] != 3],
    groups=df[df['block'] != 3]["participant"],
    re_formula="~camera"
)

result_accuracy = model_accuracy.fit()
print(result_accuracy.summary())
```

```python
model_efficiency = smf.mixedlm(
    "efficiency ~ camera",
    data=df[df['block'] != 3],
    groups=df[df['block'] != 3]["participant"],
    re_formula="~camera"
)

result_efficiency = model_efficiency.fit()
print(result_efficiency.summary())
```

```python
model_learning_accuracy = smf.mixedlm(
    "accuracy ~ camera * shape_trial_idx",
    data=df,
    groups=df["participant"],
    re_formula="~shape_trial_idx + camera"
)

result_learning_accuracy = model_learning_accuracy.fit()
print(result_learning_accuracy.summary())
```

```python
model_learning_efficiency = smf.mixedlm(
    "efficiency ~ camera * shape_trial_idx",
    data=df,
    groups=df["participant"],
    re_formula="~shape_trial_idx + camera"
)

result_learning_efficiency = model_learning_efficiency.fit()
print(result_learning_efficiency.summary())
```

```python
# Trade-off correlation
def corr_report(data, x, y, label=""):
    sub = data[[x, y]].dropna()
    n = len(sub)
    if n < 3:
        print(f"{label}{x} vs {y}: not enough data (n={n})")
        return
    r_s, p_s = spearmanr(sub[x], sub[y])
    print(f"{label}{x} vs {y} (n={n})")
    print(f"  rho={r_s:.3f}, p={p_s:.4g}")

agg_cam = (
    df
    .groupby(["participant", "camera"], as_index=False)
    .agg(
        duration_mean=("log_d", "mean"),
        chamfer_mean=("log_c", "mean"),
    )
)

for cam, g in agg_cam.groupby("camera"):
    corr_report(g, "duration_mean", "chamfer_mean", label=f"[{cam}] ")
```

```python
# Performance summary stats
def kde_peak(x):
    kde = gaussian_kde(x)
    xs = np.linspace(x.min(), x.max(), 1000)
    return xs[kde(xs).argmax()]

perf = df[df["block"] != 3]
g = perf.groupby("camera")["accuracy"]
df_summary = g.describe()[["count", "mean", "std", "min", "25%", "50%", "75%", "max"]]

df_summary['kde_peak'] = g.apply(kde_peak)
df_summary['kurtosis_excess'] = g.apply(lambda y: kurtosis(y, fisher=True, bias=False))
df_summary['skewness'] = g.apply(lambda y: skew(y, bias=False))

print("\nSummary statistics for accuracy by camera:")
display(df_summary)


perf = df[df["block"] != 3]
g = perf.groupby("camera")["efficiency"]
df_summary = g.describe()[["count", "mean", "std", "min", "25%", "50%", "75%", "max"]]

df_summary['kde_peak'] = g.apply(kde_peak)
df_summary['kurtosis_excess'] = g.apply(lambda y: kurtosis(y, fisher=True, bias=False))
df_summary['skewness'] = g.apply(lambda y: skew(y, bias=False))

print("\nSummary statistics for efficiency by camera:")
display(df_summary)
```

*Questionnaire vs. Preference*

```python
def compute_preferred_nonpreferred(row):
    preferred_score = row["stationary_cam"] if row["preferred_camera"] == "stationary_cam" else row["dynamic_cam"]
    nonpreferred_score = row["dynamic_cam"] if row["preferred_camera"] == "stationary_cam" else row["stationary_cam"]
    return pd.Series({"preferred_score": preferred_score, "nonpreferred_score": nonpreferred_score})
```

```python
# Preferred vs nonpreferred camera scores
df_questionnaire_sum = df_questionnaire.groupby(["participant", "camera"])["Response"].sum().reset_index()
df_questionnaire_sum = df_questionnaire_sum.pivot_table(index="participant", columns="camera", values="Response")

df_questionnaire_combine = df_questionnaire_sum.merge(
    df_questionnaire_preferences[df_questionnaire_preferences['shape'] == "In General"][["participant", "preferred_camera"]],
    on="participant"
)

df_questionnaire_combine[["preferred_score", "nonpreferred_score"]] = df_questionnaire_combine.apply(compute_preferred_nonpreferred, axis=1)

stat, p = wilcoxon(
    df_questionnaire_combine["preferred_score"],
    df_questionnaire_combine["nonpreferred_score"],
    alternative="greater"   # tests preferred > nonpreferred
)

print("Wilcoxon statistic:", stat)
print("p-value:", p)

preferred = sum(df_questionnaire_combine["preferred_score"] >= df_questionnaire_combine["nonpreferred_score"])
print(f"Number of participants with preferred score greater than or equal to nonpreferred score: {preferred} ({preferred/len(df_questionnaire_combine)*100:.1f}%)")
```

```python
# Preferred vs nonpreferred camera category scores
category_scores = (
    df_questionnaire
    .groupby(["participant", "camera", "category"])["Response"]
    .mean()
    .reset_index()
)

category_pivot = category_scores.pivot_table(
    index=["participant", "category"],
    columns="camera",
    values="Response"
).reset_index()

category_pivot = category_pivot.merge(
    df_questionnaire_preferences[df_questionnaire_preferences['shape'] == "In General"][["participant", "preferred_camera"]],
    on="participant"
)

category_pivot[["preferred_score", "nonpreferred_score"]] = category_pivot.apply(compute_preferred_nonpreferred, axis=1)

for cat in category_pivot["category"].unique():
    subset = category_pivot[category_pivot["category"] == cat]
    
    stat, p = wilcoxon(subset["preferred_score"], subset["nonpreferred_score"], alternative="greater")
    
    preferred = sum(subset["preferred_score"] >= subset["nonpreferred_score"])
    print(f"{cat}: Wilcoxon p-value = {p:.4f}, preferred >= nonpreferred in {preferred}/{len(subset)} participants ({preferred/len(subset)*100:.1f}%)")
```

**SUBJECTIVE VS OBJECTIVE**

```python
def compute_preference_summaries(df, df_questionnaire_preferences, metric):
    df_preference_general = (
    df_questionnaire_preferences
    .loc[
        df_questionnaire_preferences['shape'] == 'In General',
        ['participant', 'preferred_camera']
    ]
    .rename(columns={'preferred_camera': 'preferred_camera_general'})
)

    df_preferred_best = df[df['block'] != 3].groupby(['participant', 'camera'])[metric].mean().reset_index()
    df_preferred_best = df_preferred_best.groupby(['participant']).apply(
        lambda x: x.loc[x[metric].idxmax()]
    ).reset_index()

    df_preferred_best = df_preferred_best.merge(df_preference_general, on=['participant'], how='left')
    df_preferred_best['used_preferred_camera'] = df_preferred_best['camera'] == df_preferred_best['preferred_camera_general']

    df_count = df_preferred_best.groupby(['used_preferred_camera', 'preferred_camera_general']).size().reset_index(name='count')
    df_count['total'] = df_count.groupby('preferred_camera_general')['count'].transform('sum')
    df_count['percentage'] = df_count['count'] / df_count['total'] * 100

    return df_count
```

```python
df_count_eff = compute_preference_summaries(df, df_questionnaire_preferences, "efficiency")
df_count_acc = compute_preference_summaries(df, df_questionnaire_preferences, "accuracy")

print("Preferred camera usage in best-efficiency block/trial:")
for camera in df_count_eff['preferred_camera_general'].unique():
	info = df_count_eff[(df_count_eff['preferred_camera_general'] == camera) & (df_count_eff['used_preferred_camera'] == True)]
	result = binomtest(info['count'].values[0], info['total'].values[0], p=0.5, alternative='greater')
	print(f"{camera.replace('_', ' ').capitalize()}: n={info['total'].values[0]}, matched={info['count'].values[0]}, p-value = {result.pvalue:.4f}")

print("\nPreferred camera usage in best-accuracy block/trial:")
for camera in df_count_acc['preferred_camera_general'].unique():
	info = df_count_acc[(df_count_acc['preferred_camera_general'] == camera) & (df_count_acc['used_preferred_camera'] == True)]
	result = binomtest(info['count'].values[0], info['total'].values[0], p=0.5, alternative='greater')
	print(f"{camera.replace('_', ' ').capitalize()}: n={info['total'].values[0]}, matched={info['count'].values[0]}, p-value = {result.pvalue:.4f}")
```

```python
def compute_preference_summaries(df, df_questionnaire_preferences, metric):
    df_preferred_best = df[df['block'] != 3].groupby(['participant', 'shape']).apply(
	lambda x: x.loc[x[metric].idxmax()]
    )
    df_preferred_best = df_preferred_best.merge(df_questionnaire_preferences, on=['participant', 'shape'], how='left')
    df_preferred_best['used_preferred_camera'] = df_preferred_best['camera'] == df_preferred_best['preferred_camera']

    # Aggregate to one match rate per participant
    participant_match = (
        df_preferred_best.groupby(['participant', 'preferred_camera'])['used_preferred_camera']
        .mean()
        .reset_index()
    )

    # Majority-vote: did this participant match more often than not?
    participant_match['matched_majority'] = participant_match['used_preferred_camera'] > 0.5

    return participant_match
```

```python
participant_match_eff = compute_preference_summaries(df, df_questionnaire_preferences, "efficiency")
participant_match_acc = compute_preference_summaries(df, df_questionnaire_preferences, "accuracy")

print("Preferred camera usage in best-efficiency block/trial (majority-vote):")
for camera in participant_match_eff['preferred_camera'].unique():
    subset = participant_match_eff[participant_match_eff['preferred_camera'] == camera]
    n_matched = subset['matched_majority'].sum()
    n_total = len(subset)
    result = binomtest(n_matched, n_total, p=0.5, alternative='greater')
    print(f"{camera.replace('_', ' ').capitalize()}: n={n_total}, matched={n_matched}, p={result.pvalue:.4f}")

print("\nPreferred camera usage in best-accuracy block/trial (majority-vote):")
for camera in participant_match_acc['preferred_camera'].unique():
    subset = participant_match_acc[participant_match_acc['preferred_camera'] == camera]
    n_matched = subset['matched_majority'].sum()
    n_total = len(subset)
    result = binomtest(n_matched, n_total, p=0.5, alternative='greater')
    print(f"{camera.replace('_', ' ').capitalize()}: n={n_total}, matched={n_matched}, p={result.pvalue:.4f}")
```
